```python
# Separar cidade e estado
def split_city_state(value):
  value = str(value).split(' | ')
  if len(value) == 1:
    value = value[0].split(' - ')
  if len(value) == 1:
    return value[0], 'nan'
  return value[0], value[1]

# Separar cidade e estado
df['Cidade'] = df['Cidade/Estado'].apply(lambda x: split_city_state(x)[0])
df['Estado'] = df['Cidade/Estado'].apply(lambda x: split_city_state(x)[1])

# Remover coluna 'Cidade/Estado'
df.drop('Cidade/Estado', axis=1, inplace=True)

# Checar alterações se foram corretas
df.groupby(['Estado', 'Cidade']).size()
```

```python
# Remover acentos e tansformar para minúsculo o nome das cidades
for idx, city_data in enumerate(cities_data):
  cities_data[idx]['Cidade min'] = unidecode.unidecode(city_data['Cidade'].lower())
```

```python
# Buscar dados da cidade
def search_city_data(value):
  city_data = [x for x in cities_data if x['Cidade min'] == unidecode.unidecode(value).lower()]
  if len(city_data) > 0:
    return city_data[0]
  return {'ID cidade IBGE':np.nan, 'UF': np.nan, 'População IBGE 2020': np.nan, 'Latitude': np.nan, 'Longitude': np.nan}

df['Latitude'] = df['Cidade'].apply(lambda x: search_city_data(x)['Latitude'])
df['Longitude'] = df['Cidade'].apply(lambda x: search_city_data(x)['Longitude'])
df['População'] = df['Cidade'].apply(lambda x: search_city_data(x)['População IBGE 2020'])
```

 É possível extrair a quantidade de funcionários de texto para intervalos, para possibilitar a classificação dos dados

```python
# Transformar o valor de texto em intervalos
def process_company_size(value):
  value = str(value).replace('funcionários', '')
  if '-' in value:
    l, r = value.split('-')
    l, r = l.replace('.', ''), r.replace('.', '')
    return pd.Interval(left=int(l), right=int(r))
  if ' de ' in value:
    r, l = value.split(' de ')
    l, r = l.replace('.', ''), r.replace('.', '')
    return pd.Interval(left=int(l), right=np.inf)
  return pd.Interval(left=0, right=0)

# Transformar intervalo no valor médio de funcionários
def company_size_to_sequence(value, base_list):
  return int((value.left + (value.right if value.right != np.inf else value.left)) / 2)

# Criar uma lista classificada dos intervalos
intervals_list = sorted(map(process_company_size, df['Tamanho da empresa'].astype("str").unique()))

# Processar a coluna tamanho da empresa
df['Tamanho da empresa old'] = df['Tamanho da empresa']
df['Tamanho da empresa'] = df['Tamanho da empresa'].apply(process_company_size)
df['Tamanho da empresa cod'] = df['Tamanho da empresa'].apply(lambda x: company_size_to_sequence(x, intervals_list))

df['Tamanho da empresa'] = df['Tamanho da empresa'].astype(str)


df['Tamanho da empresa'] = df['Tamanho da empresa'].replace({'(10001.0, inf]': '+10001',
                                                              '(1001.0, 5000.0]': '1000 a 5000',
                                                              '(201.0, 500.0]': '200 a 500',
                                                              '(5001.0, 10000.0]': '5000 a 10000',
                                                              '(51.0, 200.0]': '51 a 200',
                                                              '(501.0, 1000.0]': '500 a 1000',
                                                              '(0.0, 0.0]': 'Sem informações',
                                                              '(2.0, 10.0]':  '2 a 10'})
```

As variáveis ​​nominais são aquelas que representam categorias sem nenhuma ordenação implícita, enquanto as variáveis ​​ordinais representam categorias com uma ordem implícita.


---



Nesse caso, os "Produto", "Tipo Saída", "Tipo Plano" e "Setor" são variáveis **​​qualitativas nominais**, pois não há ordenação implícita nas categorias que representam. 

Já as colunas "Cidade" e "Estado" são variáveis ​​**qualitativas ordinais**, pois há uma ordem implícita nas categorias que representam (por exemplo, as cidades podem ser ordenadas alfabeticamente ou por região geográfica).

são variáveis ​​qualitativas, pois representam categorias ou atributos que não podem ser medidos numericamente. 





```python
# Depois disso,  "Produto", "Tipo Saída", "Tipo Plano" e "Setor" serão transformadas em variáveis ​​nominais, onde os valores serão representados por números inteiros.
df_Numerica["Produto"] = df_Numerica["Produto"].factorize()[0]
df_Numerica["Tipo Saída"] = df_Numerica["Tipo Saída"].factorize()[0]
df_Numerica["Tipo Plano"] = df_Numerica["Tipo Plano"].factorize()[0]
df_Numerica["Setor"] = df_Numerica["Setor"].factorize()[0]
```
